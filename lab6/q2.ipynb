### Q5

```python
df['Age'].value_counts().plot(kind='bar')
```

```python
print(df['Age'].min())
```

```python
print(df['Age'].max())
```

```python
k = math.ceil(math.log(df.shape[0],2)+1)
print(f"The number of bins using Sturges' Rule is {k}")
```

```python
m = math.ceil(math.sqrt(df.shape[0]))
print(f"The number of bins using Square Root Rule is {m}")
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Age'], bins=k,edgecolor = 'Red',color='blue')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Age'], bins=m,edgecolor = 'Red',color='blue')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.grid(True)
plt.show()
```
