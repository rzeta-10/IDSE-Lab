### Q1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('shopping_trends.csv')
df
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.rename(columns={'Item Purchased':'Item_Purchased','Review Rating':'Review_Rating'}, inplace=True)
print(len(df.Item_Purchased.unique()))
```

```python
grouped_data = df.groupby('Item_Purchased')['Review_Rating'].mean().reset_index()

# Extract items and their corresponding average review ratings
items = grouped_data['Item_Purchased']
values = grouped_data['Review_Rating']

# Plotting
fig = plt.figure(figsize=(30, 10))
plt.bar(items, values)
plt.xlabel('Items')
plt.ylabel('Average Review Ratings')
plt.title('Items vs Average Review Ratings')
plt.show()
```

```python
df
```

```python
df['Review_Rating'].value_counts().sort_index(ascending=True).plot(kind='bar')
plt.xlabel('Review Ratings')
plt.ylabel('Frequency')
plt.title('Review Ratings vs Frequency')
plt.show()
```
